# Simple Test Score Predictor

### This project implements a simple linear regression model to predict student test scores based on the number of hours studied. It's designed as an introductory project for understanding how linear regression works with small datasets.

## 1) Import Required Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

## 2) Arrange Data

```python
# Input data: hours studied and corresponding test scores
time_studied = np.array([20,50,32,65,23,43,10,5,22,35,29,5,56]).reshape(-1,1)
scores = np.array([56,83,47,93,47,82,45,23,55,67,57,4,89]).reshape(-1,1)
```

## 3) Train Linear Regression Model

```python
# Split data into training and testing sets (80% train, 20% test)
time_train, time_test, scores_train, scores_test = train_test_split(time_studied, scores, test_size=0.2)
```

```python
# Initialize and train the linear regression model
model = LinearRegression()
model.fit(time_train, scores_train)
```

### R² score tells you how well your linear regression model fits the data. 1 means the model predicts perfectly (beware of overfiting), and 0 means the model is no better than just guessing.

```python
# Evaluate the model using the R² score
score = model.score(time_test, scores_test)
print(f"Model R² Score: {score:.2f}")
```

## 4) Visualize Data

### The plot  shows the relationship between hours studied and test scores. The line shows a clear positive correlation. As hours studied increases, test scores also tend to increase.

```python
# Visualize training data and the regression line
plt.figure(figsize=(8, 5))
plt.scatter(time_train, scores_train, label="Training Data", color="blue")
plt.plot(
    np.linspace(0, 70, 100).reshape(-1, 1),
    model.predict(np.linspace(0, 70, 100).reshape(-1, 1)),
    color="red",
    label="Regression Line"
)
plt.xlabel("Hours Studied")
plt.ylabel("Test Score")
plt.title("Linear Regression: Predicting Test Scores")
plt.legend()
plt.ylim(0, 100)
plt.grid(True)
plt.tight_layout()
plt.show()
```
